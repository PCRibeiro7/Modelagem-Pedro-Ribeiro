```python
# Esta c√©lula importa os pacotes que iremos utilizar
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline 
```

```python
t=sp.linspace(0,20,1000)
```

```python
v0=0
x0=0.1
wn=2*2*sp.pi
wf=3*2*sp.pi
f0=1
```

```python
def f(wf):
    x = (v0/wn)*sp.sin(wn*t) +(x0 -(f0/(wn**2-wf**2)))*sp.cos(wn*t) +(f0/(wn**2 - wf**2))*sp.cos(wf*t)
    plt.plot(t,x)
```

```python
interact(f,wf=3)
```
